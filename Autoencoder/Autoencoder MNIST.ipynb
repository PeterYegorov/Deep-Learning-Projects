```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.svm import SVC
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

# Load and preprocess MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
x_train = x_train.reshape((-1, 784))
x_test = x_test.reshape((-1, 784))

def build_autoencoder(hidden_units):
    input_layer = Input(shape=(784,))
    encoded = Dense(hidden_units, activation='relu')(input_layer)
    decoded = Dense(784, activation='sigmoid')(encoded)
    autoencoder = Model(input_layer, decoded)
    autoencoder.compile(optimizer='adam', loss='mean_squared_error')
    return autoencoder

# 1. Investigate the number of hidden units on reconstruction performance
hidden_units_list = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60]
classifiers = {}
accuracies = []
reconstruction_losses = []

for hidden_units in hidden_units_list:
    autoencoder = build_autoencoder(hidden_units)
    autoencoder.fit(x_train, x_train, epochs=10, batch_size=256, shuffle=True)

    # Plot loss function
    plt.plot(autoencoder.history.history['loss'])
    plt.title(f'Reconstruction Loss for {hidden_units} Hidden Units')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.show()

    # Display the set of weights as an image
    if hidden_units in [5, 10]:
        encoder_weights = autoencoder.layers[1].get_weights()[0]
        fig, axes = plt.subplots(1, hidden_units, figsize=(10, 2))
        for i in range(hidden_units):
            axes[i].imshow(encoder_weights[:, i].reshape(28, 28), cmap='gray')
            axes[i].axis('off')
        plt.suptitle(f'Weights for {hidden_units} Hidden Units')
        plt.show()
    
    encoded_x_train = autoencoder.layers[1](x_train)
    encoded_x_test = autoencoder.layers[1](x_test)
    
    # Train a classifier (e.g., SVM) on the encoded data
    clf = SVC()
    clf.fit(encoded_x_train, y_train)
    y_pred = clf.predict(encoded_x_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)
    print(f"Hidden Units: {hidden_units}, Accuracy: {accuracy * 100:.2f}%")

    # Generate confusion matrix
    cm = confusion_matrix(y_test, y_pred, normalize='true')
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    plt.imshow(cm_normalized, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix (Hidden Units: {hidden_units})')
    plt.colorbar()
    tick_marks = np.arange(10)
    plt.xticks(tick_marks, range(10))
    plt.yticks(tick_marks, range(10))
    
    for i in range(10):
        for j in range(10):
            plt.text(j, i, f'{cm[i, j]:.2f}', horizontalalignment="center", color="white" if cm[i, j] > cm.max() / 2 else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()

    # Collect reconstruction loss
    reconstruction_losses.append(autoencoder.history.history['loss'][-1])

# Plot testing performance vs the number of hidden neurons
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(hidden_units_list, accuracies, marker='o')
plt.title('Testing Performance vs Hidden Units')
plt.xlabel('Number of Hidden Units')
plt.ylabel('Accuracy')

# Plot reconstruction performance vs the number of hidden neurons
plt.subplot(1, 2, 2)
plt.plot(hidden_units_list, reconstruction_losses, marker='o', color='orange')
plt.title('Reconstruction Performance vs Hidden Units')
plt.xlabel('Number of Hidden Units')
plt.ylabel('Reconstruction Loss')

plt.tight_layout()
plt.show()
```
